# Manual Face Detection

```python
file_path = os.path.join(RESULTS_PATH, "manual_face_detection", "manual_validation_raw.csv")
df_manual = pd.read_csv(file_path)
df_manual = df_manual.iloc[:, 1:]
```

```python
df_manual["precision"] = df_manual["true_positive"] / (df_manual["true_positive"] + df_manual["false_positive"])
df_manual["recall"] = df_manual["true_positive"] / (df_manual["true_positive"] + df_manual["false_negative"])
df_manual["f1_score"] = 2 * df_manual["precision"] * df_manual["recall"] / (df_manual["precision"] + df_manual["recall"])
df_manual["precision"] = df_manual["precision"] * 100.0
df_manual["recall"] = df_manual["recall"] * 100.0
df_manual["f1_score"] = df_manual["f1_score"] * 100.0
df_manual = df_manual.sort_values("detector")
df_manual = df_manual.iloc[:, [0, 1, 5, 6, 7]]
```

```python
df_manual = df_manual.round({"precision": 2, "recall": 2, "f1_score": 2})
df_manual = df_manual.rename({"detector": "Detektor",
                              "equal_hist": "Wyrów. hist.",
                              "precision": "Precision",
                              "recall": "Recall",
                              "f1_score": "F1 score"
                             }, axis=1)
```

```python
df_manual
```

```python
latex = df_manual.to_latex(index=False).replace(".", ",").replace("True", "tak").replace("False", "nie")
with open(os.path.join(FOR_THESIS_DIR, "manual_face_detection.txt"), 'w') as file:
    file.write(latex)
```
